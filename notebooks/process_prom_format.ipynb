```python
# warnings....ignore
import pandas as pd
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)


pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv("../statics/csv/cc_prom.csv")
df.head()
```

```python
# Exact columns
samples = df.columns.to_list()[1:-1]
print("Columns sample: ", samples)
sample_group = df.groupby(samples)[samples].agg(['unique']).reset_index().droplevel(1, axis=1)
print(f"Have {len(sample_group)} samples")
```

```python
df['Time'] = pd.to_datetime(df["Time"], unit="ms")
df.head()
buckets = df.drop_duplicates(subset="le")['le'].to_list()
instances = df.groupby(samples).last().reset_index()
print("List bucket for bins: ", buckets)

```

```python
# rename columns
cols = {
    "Time": "timestamp",
    "Value #A": "value"
}
df = df.rename(columns=cols)
df.head()
```

```python
df_grouped = df.groupby([pd.Grouper(freq="5min", key="timestamp")] + samples).first().reset_index()
df_grouped.head()
print("Total row: ", len(df_grouped))
df_grouped.head()
```

```python
# Implement rate calculation for each group
list_df_rated = []


def rate(group: pd.Series):
    sample = df_grouped[(df_grouped["__name__"] == group['__name__'])
                        & (df_grouped["app"] == group['app'])
                        & (df_grouped["controller_class"] == group['controller_class'])
                        & (df_grouped["controller_namespace"] == group['controller_namespace'])
                        & (df_grouped["controller_pod"] == group['controller_pod'])
                        & (df_grouped["host"] == group['host'])
                        & (df_grouped["ingress"] == group['ingress'])
                        & (df_grouped["instance"] == group['instance'])
                        & (df_grouped["job"] == group['job'])
                        & (df_grouped["kubernetes_namespace"] == group['kubernetes_namespace'])
                        & (df_grouped["kubernetes_node"] == group['kubernetes_node'])
                        & (df_grouped["kubernetes_pod_name"] == group['kubernetes_pod_name'])
                        & (df_grouped["le"] == group['le'])
                        & (df_grouped["method"] == group['method'])
                        & (df_grouped["namespace"] == group['namespace'])
                        & (df_grouped["ns"] == group['ns'])
                        & (df_grouped["path"] == group['path'])
                        & (df_grouped["pod_template_hash"] == group['pod_template_hash'])
                        & (df_grouped["service"] == group['service'])
                        & (df_grouped["status"] == group['status'])].reset_index(drop=True)
    sample['rate'] = sample['value'].diff(
    ) / sample['timestamp'].diff().dt.total_seconds()
    rate_index_2 = sample['rate'][sample.index == 1]
    sample.at[0, 'rate'] = rate_index_2
    sample.fillna(0, inplace=True)
    list_df_rated.append(sample)


for idx, group in instances.iterrows():
    rate(group)

df_result = pd.concat(list_df_rated)
df_result.head(5)
```
